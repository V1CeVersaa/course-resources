```python
import torch


def test_meshgrid():
    idx_1 = torch.arange(4)
    idx_2 = torch.arange(3) + 5
    idx_x, idx_y = torch.meshgrid(idx_1, idx_2, indexing="ij")
    stacked = torch.stack([idx_x.flatten(), idx_y.flatten()], dim=1)
    print(idx_x)
    print(idx_y)
    print(stacked)

    # standard way
    idx_x, idx_y = torch.meshgrid(idx_1, idx_2, indexing="xy")
    stacked = torch.stack([idx_x.flatten(), idx_y.flatten()], dim=1)
    print(idx_x)
    print(idx_y)
    print(stacked)


test_meshgrid()
```

```python
def test_unbind():
    x = torch.tensor([[1, 2], [1, 2], [1, 2]])
    x0, x1 = x.unbind(dim=1)
    print(x0)
    print(x1)
    stacked = torch.stack([x0, x1], dim=1)
    print(stacked)


test_unbind()
```
